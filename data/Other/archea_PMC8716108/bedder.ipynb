```python
import pandas as pd
```

```python
df = pd.read_csv("tts.csv")
df
```

```python
df["score"] = df["S2_head.reads"].astype(int) + df["▽aCPSF1_head.reads"].astype(int)
df.drop(["S2_head.reads", "▽aCPSF1_head.reads"], axis=1, inplace=True)
df.head()
```

```python
df["start"] = df["Position"] - 1
df["end"] = df["Position"]

df.drop(["Position",], axis=1, inplace=True)
```

```python
df
```

```python
df["gene"] = [f"NC_005791.1_{i}" for i in range(len(df))]

df = df [["gene", "start", "end", "TTS categories", "score", "Strand"]]

df 
```

```python
df.isna().sum()
```

```python
# sort by start
df.sort_values(by=["start"], inplace=True)
```

```python
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    

df.to_csv('MethanococcusMaripaludis_S2_NC005791.1.bed', sep='\t', index=False, header=False)
```
