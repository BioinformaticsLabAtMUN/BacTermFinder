```python
import pandas as pd
# show all columns 
pd.set_option('display.max_columns', None)
```

```python
ecoli = pd.read_excel("mbo004141900st3.xlsx", sheet_name="table S3 Terminators", header=1, skipfooter=0)
ecoli
```

```python
ecoli = ecoli[["Term", "Strand", "gene", ]]
```

```python
ecoli["Term"] = ecoli["Term"].apply(lambda x: x.split("-")[1])
ecoli["Term"] = ecoli["Term"].astype(int)
```

```python
ecoli['start'] = ecoli['Term'] - 1
ecoli['end'] = ecoli['Term']


ecoli.drop(columns=['Term'], inplace=True)

ecoli['ref_gene'] = [f"U00096.2_{i}" for i in range(len(ecoli))]
```

```python
ecoli.columns
```

```python
ecoli["score"] = 0
ecoli = ecoli[["ref_gene", "start", "end", "gene", "score", "Strand"]]
ecoli
```

```python
ecoli.to_csv("EscherichiaColi_K12-MG1655_U00096.2.bed", sep='\t', header=False, index=False)
```
