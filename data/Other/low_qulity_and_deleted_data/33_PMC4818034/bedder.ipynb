```python
import pandas as pd
# show all columns 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel("pgen.1005962.s009.xlsx", sheet_name="Table S7", header=1, skipfooter=0)
df
```

```python
df = df[["Strand", "Pos", "MaxEfRhoDel", "Id"]]
df
```

```python
df["gene"] = [f"CP000253.1_{i}" for i in range(len(df))]

# df.drop(columns=["Start_TT"], inplace=True)
df
```

```python
df["start"] = df["Pos"] - 1
df["end"] = df["Pos"]
df.drop(columns=["Pos"], inplace=True)
df
```

```python
df["strand"] = df["Strand"].apply(lambda x: "+" if x == 1 else "-")
```

```python
df = df[["gene", "start", "end", "Id", "MaxEfRhoDel", "strand"]]
df
```

```python
df.to_csv("StaphylococcusAureus_NCTC8325_CP000253.1.bed", sep='\t', header=False, index=False)
```
