```python
import pandas as pd
# show all columns 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel("1471-2164-13-567-S15.xlsx", sheet_name="5-UTR", header=2275, skipfooter=0)
df
```

```python
df = df[["Start", "End", "Strand", "Putative transcriptional unit", "3&apos;-UTR length"]]
df
```

```python
# creating a tts column if the strand is + End is the tts else Start is the tts
df["tts"] = df.apply(lambda x: x["End"] if x["Strand"] == "+" else x["Start"], axis=1)
df.drop(["Start", "End", ], axis=1, inplace=True)
df["start"] = df["tts"] - 1
df["end"] = df["tts"] 
df.drop(["tts"], axis=1, inplace=True)
df
```

```python
# mask that has "PBPRA" in it
mask = df["Putative transcriptional unit"].str.contains("PBPRA")
chr1 = df[mask]

mask = df["Putative transcriptional unit"].str.contains("PBPRB")
chr2 = df[mask]

mask = df["Putative transcriptional unit"].str.contains("PBPRC")
plsmd = df[mask]

print(len(chr1) + len(chr2) + len(plsmd))
```

```python
chr1["gene"] = [f"NC_006370.1_{i}" for i in range(len(chr1))]
chr2["gene"] = [f"NC_006371.1_{i}" for i in range(len(chr2))]
plsmd["gene"] = [f"NC_005871.1_{i}" for i in range(len(plsmd))]
```

```python
chr1 = chr1[["gene", "start", "end", "Putative transcriptional unit", "3&apos;-UTR length", "Strand"]]
chr1
```

```python
chr2 = chr2[["gene", "start", "end", "Putative transcriptional unit", "3&apos;-UTR length", "Strand"]]
chr2
```

```python
plsmd = plsmd[["gene", "start", "end", "Putative transcriptional unit", "3&apos;-UTR length", "Strand"]]
plsmd
```

```python
chr1.to_csv("PhotobacteriumProfundum_SS9(chr1)_NC006370.1.bed", sep='\t', header=False, index=False)
chr2.to_csv("PhotobacteriumProfundum_SS9(chr2)_NC006371.1.bed", sep='\t', header=False, index=False)
plsmd.to_csv("PhotobacteriumProfundum_SS9(plsmd)_NC005871.1.bed", sep='\t', header=False, index=False)
```
