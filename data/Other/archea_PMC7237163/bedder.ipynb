```python
import pandas as pd
```

```python
df = pd.read_csv("tts.csv")
df
```

```python
df["start"] = df["TTS"].astype(int) - 1
df["end"] = df["TTS"].astype(int) 

df.drop(columns=["TTS"], inplace=True)

df
```

```python
df["#chr"].value_counts()
```

```python
CHR = df[df["#chr"] == "CHR"].copy ()
phv4 = df[df["#chr"] == "pHV4"].copy ()
phv3 = df[df["#chr"] == "pHV3"].copy ()
phv1 = df[df["#chr"] == "pHV1"].copy ()
```

```python
CHR["gene"] = [f"CHR_{i}" for i in range(len(CHR))]
phv4["gene"] = [f"pHV4_{i}" for i in range(len(phv4))]
phv3["gene"] = [f"pHV3_{i}" for i in range(len(phv3))]
phv1["gene"] = [f"pHV1_{i}" for i in range(len(phv1))]

# CHR.drop(columns=["#chr"], inplace=True)
# phv4.drop(columns=["#chr"], inplace=True)
# phv3.drop(columns=["#chr"], inplace=True)
# phv1.drop(columns=["#chr"], inplace=True)

CHR = CHR [["gene", "start", "end", "#chr", "usTTS", "strand"]]
phv4 = phv4 [["gene", "start", "end", "#chr", "usTTS", "strand"]]
phv3 = phv3 [["gene", "start", "end", "#chr", "usTTS", "strand"]]
phv1 = phv1 [["gene", "start", "end", "#chr", "usTTS", "strand"]]

print(CHR.head())
print(phv4.head())
print(phv3.head())
print(phv1.head())
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    

CHR.to_csv('HaloferaxVolcanii_H119(CHR)_XXXXXX.bed', sep='\t', index=False, header=False)
phv4.to_csv('HaloferaxVolcanii_H119(pHV4)_XXXXXX.bed', sep='\t', index=False, header=False)
phv3.to_csv('HaloferaxVolcanii_H119(pHV3)_XXXXXX.bed', sep='\t', index=False, header=False)
phv1.to_csv('HaloferaxVolcanii_H119(pHV1)_XXXXXX.bed', sep='\t', index=False, header=False)

```
