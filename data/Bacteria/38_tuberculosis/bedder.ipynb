```python
import pandas as pd
# show all columns 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel("media-4.xlsx", sheet_name="Classification of TTS", header=0, skipfooter=0)
df
```

```python
# select only I and A from class
df = df[df['class'].isin(['I', 'A'])]
df
```

```python
df = df[["TTS position", "score", "strand", ]]
```

```python
# df = df[df["Is mRNA?"] == 1]
df 
```

```python
# df['start'] = df[df['Strand'] == 0 ] ["5' TSS"] - 1
# df['end'] = df[df['Strand'] == 0 ] ["5' TSS"]

df['start'] = df["TTS position"] - 1
df['end'] = df["TTS position"]

df['ref_gene'] = [f"AL123456.3" for i in range(len(df))]
```

```python
df
```

```python
df['name'] = "name" 
# df.drop(columns=["ref_gene"], inplace=True)
df = df[["ref_gene", "start", "end", 'name', "score", "strand"]]
df['start'] = df['start'].astype(int)
df['end'] = df['end'].astype(int)
```

```python
df
```

```python
df.to_csv("MycobacteriumTuberculosis_H37Rv_AL123456.3.bed", sep='\t', header=False, index=False)
```
