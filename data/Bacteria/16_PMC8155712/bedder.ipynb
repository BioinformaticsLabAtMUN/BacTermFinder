```python
import pandas as pd
```

```python
df = pd.read_excel("Data_Sheet_2.xlsx", sheet_name= "Sheet1", header = 1)

df
```

```python
df.drop(["TEP ID",], axis = 1, inplace = True)
```

```python
df = df[(df["Category"] == "P") | (df["Category"] == "S")]
df.drop(["Category"], axis = 1, inplace = True)
df
```

```python
df.reset_index(drop = True, inplace = True)

df
```

```python
df['start'] = df['Position'] - 1
df['end'] = df['Position']

df.drop(["Position"], axis = 1, inplace = True)

df["score"] = 0

df
```

```python
df["Locus"] = df["Locus"]+[".1"]
df
```

```python
df = df [["Locus", "start", "end", "Associated gene", "score", "Strand"]]
```

```python
# split into different files based on locus
loci = df["Locus"].unique()
for locus in loci:
    df[df["Locus"] == locus].to_csv("Synechocystis_PCC6803_"+locus+".bed", sep = "\t", header = False, index = False)
```
