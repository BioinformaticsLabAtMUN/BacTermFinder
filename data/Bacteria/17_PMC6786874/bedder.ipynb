```python
import pandas as pd
```

```python
df = pd.read_excel("mBio.02253-19-st001.xlsx", sheet_name= "S1A - TTS associated with genes", header = 2)

df
```

```python
df = df [["Strand", "TTS position", "TTS counts", "Locus"]]

df
```

```python
df.dropna(inplace = True)
df
```

```python
df["TTS position"] =  df["TTS position"].astype(int)
df["TTS counts"] = df["TTS counts"].astype(int)
df
```

```python
df.reset_index(drop = True, inplace = True)

df["gene"] = ["NC_002516.2_{}".format(i) for i in range(len(df))]
df
```

```python
df['start'] = df['TTS position'] - 1
df['end'] = df['TTS position']

df.drop(["TTS position"], axis = 1, inplace = True)

df["score"] = df["TTS counts"]

df
```

```python
df.drop(["TTS counts"], axis = 1, inplace = True)
df
```

```python
df = df [["gene", "start", "end", "Locus", "score", "Strand"]]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('PseudomonasAeruginosa_PAO1_NC002516.2.bed', sep='\t', index=False, header=False)
```
