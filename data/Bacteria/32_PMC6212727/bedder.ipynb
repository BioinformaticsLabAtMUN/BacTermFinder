```python
import pandas as pd
# show all columns 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel("Supplementary Tables_revision2_v1.xlsx", sheet_name="Table S8 - Terminators", header=0, skipfooter=1)
df
```

```python
df = df[["Coordinates", "Termination efficiency (%)", "Hairpin energy"]]
df
```

```python
df["gene"] = [f"CP027540.1_{i}" for i in range(len(df))]

# df.drop(columns=["Start_TT"], inplace=True)
df
```

```python
# split the coordinates column into 3 columns
df["start"] = df["Coordinates"].apply(lambda x: x.split("-")[0],)
df["end"] = df["Coordinates"].apply(lambda x: x.split("-")[1].split(" ")[0],)
df["strand"] = df["Coordinates"].apply(lambda x: x.split(" ")[1][1],)
df
```

```python
df
```

```python
df["start"] = df["start"].astype(int)
df["end"] = df["end"].astype(int)
```

```python
# df["tts"] = df.apply(lambda x: x["end"] if x["strand"] == "+" else x["start"], axis=1)
# writing above line in a more readable way
def get_tts(row):
    if row["strand"] == "+":
        return row["end"]
    else:
        return row["start"]
df["tts"] = df.apply(get_tts, axis=1)

df["start"] = df["tts"] - 1
df["end"] = df["tts"]
df
```

```python
df = df[["gene", "start", "end", "Hairpin energy", "Termination efficiency (%)", "strand"]]
df
```

```python
df.to_csv("StreptococcusPneumoniae_D39V_CP027540.1.bed", sep='\t', header=False, index=False)
```
