```python
import pandas as pd
# show all columns 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel("gb-2011-12-5-r47-S1 (1).XLSX", sheet_name="Table S1", header=0, skipfooter=0)
df
```

```python
df = df[["3' End", "5' TSS", "Strand", "Is mRNA?", "Mean RNA sequencing over transcript", ]]
```

```python
# df = df[df["Is mRNA?"] == 1]
df 
```

```python
def get_tts(row):
    if row["Strand"] == 1:
        return row["3' End"]
    elif row["Strand"] == 0:
        return row["3' End"]

df['tts'] = df.apply(get_tts, axis=1)
```

```python
df
```

```python
# df['start'] = df[df['Strand'] == 0 ] ["5' TSS"] - 1
# df['end'] = df[df['Strand'] == 0 ] ["5' TSS"]

df['start'] = df["tts"] - 1
df['end'] = df["tts"]


df.drop(columns=["3' End", "5' TSS"], inplace=True)

df['ref_gene'] = [f"CP000100.1" for i in range(len(df))]
```

```python

df["score"] = df["Mean RNA sequencing over transcript"]
df.drop(columns = "Mean RNA sequencing over transcript", inplace=True)

df
```

```python
df["Strand"] = df["Strand"].apply(lambda x: '+' if x==1 else '-')
```

```python
df['name'] = "name" 
# df.drop(columns=["ref_gene"], inplace=True)
df = df[["ref_gene", "start", "end", 'name', "score", "Strand"]]
df['start'] = df['start'].astype(int)
df['end'] = df['end'].astype(int)
```
