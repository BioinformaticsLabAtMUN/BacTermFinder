```python
import pandas as pd
```

```python
df = pd.read_excel("41467_2022_31177_MOESM4_ESM.xlsx", sheet_name= "Sup Tab 1. TSS and Terminators Table", header = 0)

df
```

```python
df = df[df["Type"] == "terminator"].copy()

df
```

```python
df.reset_index(inplace = True)
df.drop(["TSS type", "libs"], inplace = True, axis = 1)
df
```

```python
df.drop(["index"], inplace = True, axis = 1)
```

```python
df.drop(["Type"], inplace = True, axis = 1)
```

```python
df
```

```python
fwd = df["strand"] == "+"
rev = df["strand"] == "-"

df["tts"] = pd.concat([df[fwd]["end"] , df[rev]["start"]], )

df.drop(["start", "end"], inplace = True, axis = 1)
df
```

```python
df["gene"] = ["LR027876.1" for _ in range(len(df))]
df
```

```python
df['start'] = df['tts'] - 1
df['end'] = df['tts']

df.drop(["tts"], axis = 1, inplace = True)

df["score"] = 0

df
```

```python
df = df [["gene", "start", "end", "associated_gene", "score", "strand"]]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('StaphylococcusAureus_JKD6009_LR027876.1.bed', sep='\t', index=False, header=False)
```
