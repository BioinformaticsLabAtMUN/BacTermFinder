```python
import pandas as pd
```

```python
df = pd.read_excel("ppat.1007461.s006 (1).xlsx", sheet_name= "TTS.10.1", header = 0)
df
```

```python
df.drop(["From", "To", "Locus", "Coverage", "Enrichment_over_background", "3&apos;UTR_length", "40nt_upstream", "20nt_downstream", "Fold",], axis = 1, inplace = True)
df
```

```python
df.drop(["No._of_Us",], axis = 1, inplace = True)
df
```

```python
df["gene"] = ["NC_003028.3" + "_{}".format(j) for j in range(len(df))]
df
```

```python
df['start'] = df['TTS'] - 1
df['end'] = df['TTS']

df.drop(["TTS"], axis = 1, inplace = True)

df
```

```python
s = pd.read_excel("ppat.1007461.s006 (1).xlsx", sheet_name= "TTS.10.1", header = 0)
df["name"] = s["Locus"].copy()

df
```

```python
df.isna().sum()
```

```python

df = df[["gene", "start", "end", "name", "MFE", "Strand"]]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('StreptococcusPneumoniae_TIGR4_NC003028.3.bed', sep='\t', index=False, header=False)
```
