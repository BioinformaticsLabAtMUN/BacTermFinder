```python
import pandas as pd
```

```python
df = pd.read_excel("mSystems.00250-20-sd003.xlsx",sheet_name="TTS_list")
df
```

```python
df.drop(["A", "B", "C", "D", "E", "F", "Name", ], axis=1, inplace=True)
df.head()
```

```python
df["start"] = df["Position"] - 1
df["end"] = df["Position"]

df.drop(["Position", "PreTerm"], axis=1, inplace=True)

df = df [["Sequence", "start", "end", "Location_class", "Strand"]]

df 
```

```python
df["score"] = 0
df
```

```python
seqs = df["Sequence"].value_counts()
```

```python
seqs
```

```python
ZM4 = df[ df["Sequence"] == seqs.index[0] ].copy()
pZM32 = df[ df["Sequence"] == seqs.index[1] ].copy()
pZM33 = df[ df["Sequence"] == seqs.index[2] ].copy()
pZM36 = df[ df["Sequence"] == seqs.index[3] ].copy()
pZM39 = df[ df["Sequence"] == seqs.index[4] ].copy()
```

```python
ZM4["gene"] = [f"CP023715.1_{i}" for i in range(len(ZM4))]
pZM32["gene"] = [f"CP023716.1_{i}" for i in range(len(pZM32))]
pZM33["gene"] = [f"CP023717.1_{i}" for i in range(len(pZM33))]
pZM36["gene"] = [f"CP023718.1_{i}" for i in range(len(pZM36))]
pZM39["gene"] = [f"CP023719.1_{i}" for i in range(len(pZM39))]

ZM4.drop(["Sequence"], axis=1, inplace=True)
pZM32.drop(["Sequence"], axis=1, inplace=True)
pZM33.drop(["Sequence"], axis=1, inplace=True)
pZM36.drop(["Sequence"], axis=1, inplace=True)
pZM39.drop(["Sequence"], axis=1, inplace=True)
```

```python
# chr.reset_index(drop=True, inplace=True)
# plsmd.reset_index(drop=True, inplace=True)
```

```python
ZM4 = ZM4[["gene", "start", "end", "Location_class", "score", "Strand"]]
pZM32 = pZM32[["gene", "start", "end", "Location_class", "score", "Strand"]]
pZM33 = pZM33[["gene", "start", "end", "Location_class", "score", "Strand"]]
pZM36 = pZM36[["gene", "start", "end", "Location_class", "score", "Strand"]]
pZM39 = pZM39[["gene", "start", "end", "Location_class", "score", "Strand"]]
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    

ZM4.to_csv('ZymomonasMobilis_ATCC31821(ZM4)_CP023715.1.bed', sep='\t', index=False, header=False)
pZM32.to_csv('ZymomonasMobilis_ATCC31821(pZM32)_CP023716.1.bed', sep='\t', index=False, header=False)
pZM33.to_csv('ZymomonasMobilis_ATCC31821(pZM33)_CP023717.1.bed', sep='\t', index=False, header=False)
pZM36.to_csv('ZymomonasMobilis_ATCC31821(pZM36)_CP023718.1.bed', sep='\t', index=False, header=False)
pZM39.to_csv('ZymomonasMobilis_ATCC31821(pZM39)_CP023719.1.bed', sep='\t', index=False, header=False)
```
