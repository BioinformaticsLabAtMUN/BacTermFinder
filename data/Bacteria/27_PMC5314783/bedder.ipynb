```python
import pandas as pd
```

```python
df = pd.read_excel("mbio.00524-22-st001.xlsx", sheet_name= "Supplementary Table S1C", header = 2)
```

```python
df
```

```python
df.drop(["TU #"], axis = 1, inplace = True)
df["gene"] = [f"NC_014500.1" for i in range(len(df))]
df["start"] = df["TTS position"] - 1
df["end"] = df["TTS position"]
df.drop(["TTS position"], axis = 1, inplace = True)
df["name"] = df["gene upstream"] + " | " + df["confidence"] + " | " + df["type"]
df["score"] = df["relative strength"].copy()
df.drop(["gene upstream", "confidence", "type", "relative strength"], axis = 1, inplace = True)

# drop nan
df.dropna(inplace = True)

df
    
```

```python
df = df[["gene", "start", "end", "name", "score", "strand"]]
df.head ()
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('DickeyaDadantii_3937_NC014500.1.bed', sep='\t', index=False, header=False)
```
