```python
import pandas as pd
```

```python
df = pd.read_excel("pnas.2103579118.sd04.xlsx", sheet_name="TTSs")
scores = pd.read_excel("pnas.2103579118.sd04.xlsx", sheet_name="folding", header=2)

df
```

```python
scores
```

```python
#  add the "Folding energy(kcal/mol)" column from the scores dataframe to df dataframe based on the TTS Position column of scores to TTS of df
df_1 = df.merge(scores, right_on="TTS Position", left_on="TTS")
df_1
```

```python
# convert 'NaN' to Nan
df_1 = df_1.replace('NaN', pd.NA)
```

```python
df_1 = df_1[(df_1["Locus tag"] != "orphan")  & (df_1["Type"] == "gene") & (df_1["Location TTS"] == "3UTR")]
df_1
```

```python
df_1.drop(columns=["Old locus tag", "Name", "Type", "Start", "End", "Location TTS", "TTS Position", "Sequence", "Secondary Structure"], inplace=True)
df_1
```

```python
# chromosome, start, end, Associated gene , score, strand
df_1['gene'] = [f"CP010905.2_{i}" for i in range(len(df_1))]
df_1["start"] = df_1["TTS"] - 1
df_1["end"] = df_1["TTS"]
df_1.drop(columns=["TTS"], inplace=True)
df_1
```

# Adding strand from gene annotation

```python
# read sequence.txt file and save the lines that have "complement" in them in a list
with open("sequence.txt", "r") as f:
    lines = f.readlines()
    lines = [line.strip() for line in lines if "complement" in line]
```

```python

# extract the locus tag of the complement regions
locus = [line.split("locus_tag=")[1].split("]")[0] for line in lines]
print(locus[1:10])
```

```python
df_1["strand"] = df_1.apply(lambda x: "+" if x["Locus tag"] not in locus else "-", axis=1)
```

```python

df_1.head(20)
```

```python
df_1 = df_1[["gene", "start", "end", "Locus tag", "Folding energy (kcal/mol)", "strand"]]
df_1
```

```python
df_1.to_csv("ClostridioidesDifficile_630_CP010905.2.bed", sep='\t', header=False, index=False)
```
