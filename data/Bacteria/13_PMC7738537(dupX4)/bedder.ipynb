```python
import pandas as pd
```

```python
df = pd.read_csv("Dataset_figshare_2.csv", header = 2)
df
```

```python
# seperate df into 7 dfs for each 4 cols
df_1 = df.iloc[:, 0:4]
df_2 = df.iloc[:, 4:8]
df_3 = df.iloc[:, 8:12]
df_4 = df.iloc[:, 12:16]
df_5 = df.iloc[:, 16:20]
df_6 = df.iloc[:, 20:24]
df_7 = df.iloc[:, 24:28]

df_1.head()
```

```python
df_1.dropna(inplace = True)
df_2.dropna(inplace = True)
df_3.dropna(inplace = True)
df_4.dropna(inplace = True)
df_5.dropna(inplace = True)
df_6.dropna(inplace = True)
df_7.dropna(inplace = True)
```

```python
df_2.columns
```

```python
df_1["genome"] = ["BA000030.4" + f"_{i}" for i in range(len(df_1))]
df_2["genome"] = ["CP027858.1" + f"_{i}" for i in range(len(df_2))]
df_3["genome"] = ["NC_003888.3" + f"_{i}" for i in range(len(df_3))]
df_4["genome"] = ["NC_010572.1" + f"_{i}" for i in range(len(df_4))]
df_5["genome"] = ["CP009124.1" + f"_{i}" for i in range(len(df_5))]
df_6["genome"] = ["CP020700.1" + f"_{i}" for i in range(len(df_6))]
df_7["genome"] = ["CP059991.1" + f"_{i}" for i in range(len(df_7))]



df_1["start"] = df_1["Position"] - 1
df_1["end"] = df_1["Position"]

df_2["start"] = df_2["Position.1"] - 1
df_2["end"] = df_2["Position.1"] 

df_3["start"] = df_3["Position.2"] - 1
df_3["end"] = df_3["Position.2"] 

df_4["start"] = df_4["Position.3"] - 1 
df_4["end"] = df_4["Position.3"] 

df_5["start"] = df_5["Position.4"] - 1
df_5["end"] = df_5["Position.4"] 

df_6["start"] = df_6["Position.5"] - 1
df_6["end"] = df_6["Position.5"] 

df_7["start"] = df_7["Position.6"] - 1
df_7["end"] = df_7["Position.6"] 

df_1
```

```python
df_1.drop(columns = ["Position"], inplace = True)
df_2.drop(columns = ["Position.1"], inplace = True)
df_3.drop(columns = ["Position.2"], inplace = True)
df_4.drop(columns = ["Position.3"], inplace = True)
df_5.drop(columns = ["Position.4"], inplace = True)
df_6.drop(columns = ["Position.5"], inplace = True)
df_7.drop(columns = ["Position.6"], inplace = True)
```

```python
df_1 = df_1 [["genome", "start", "end", "Reference genome", "z-score", "Strand"]]
df_2 = df_2 [["genome", "start", "end", "Reference genome.1", "z-score.1", "Strand.1"]]
df_3 = df_3 [["genome", "start", "end", "Reference genome.2", "z-score.2", "Strand.2"]]
df_4 = df_4 [["genome", "start", "end", "Reference genome.3", "z-score.3", "Strand.3"]]
df_5 = df_5 [["genome", "start", "end", "Reference genome.4", "z-score.4", "Strand.4"]]
df_6 = df_6 [["genome", "start", "end", "Reference genome.5", "z-score.5", "Strand.5"]]
df_7 = df_7 [["genome", "start", "end", "Reference genome.6", "z-score.6", "Strand.6"]]

df_1
```

```python
# change the type of columns start and end to int
df_1["start"] = df_1["start"].astype(int)
df_1["end"] = df_1["end"].astype(int)
df_2["start"] = df_2["start"].astype(int)
df_2["end"] = df_2["end"].astype(int)
df_3["start"] = df_3["start"].astype(int)
df_3["end"] = df_3["end"].astype(int)
df_4["start"] = df_4["start"].astype(int)
df_4["end"] = df_4["end"].astype(int)
df_5["start"] = df_5["start"].astype(int)
df_5["end"] = df_5["end"].astype(int)
df_6["start"] = df_6["start"].astype(int)
df_6["end"] = df_6["end"].astype(int)
df_7["start"] = df_7["start"].astype(int)
df_7["end"] = df_7["end"].astype(int)

```

```python
df_7
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    

df_1.to_csv('StreptomycesAvermitilis_MA4680_BA000030.4.bed', sep='\t', index=False, header=False)
df_2.to_csv('StreptomycesClavuligerus_ATCC27064_CP027858.1.bed', sep='\t', index=False, header=False)
df_3.to_csv('StreptomycesCoelicolor_M145_NC003888.3.bed', sep='\t', index=False, header=False)
df_4.to_csv('StreptomycesGriseus_NBRC13350_NC010572.1.bed', sep='\t', index=False, header=False)
df_5.to_csv('StreptomycesLividans_TK24_CP009124.1.bed', sep='\t', index=False, header=False)
df_6.to_csv('StreptomycesTsukubaensis_NBRC108819_CP020700.1.bed', sep='\t', index=False, header=False)
df_7.to_csv('StreptomycesVenezuelae_ATCC15439_CP059991.1.bed', sep='\t', index=False, header=False)
```
