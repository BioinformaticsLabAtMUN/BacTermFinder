## Get the read through terminators

```python
import pandas as pd

bsub = pd.read_excel("NIHMS964513-supplement-Table_S3.xlsx", sheet_name= "5_tuned_term_Bsub", header = 7)
ecol = pd.read_excel("NIHMS964513-supplement-Table_S3.xlsx", sheet_name= "6_tuned_term_Ecol", header = 7)
vnat = pd.read_excel("NIHMS964513-supplement-Table_S3.xlsx", sheet_name= "7_tuned_term_Vnat", header = 7)
ccre = pd.read_excel("NIHMS964513-supplement-Table_S3.xlsx", sheet_name= "8_tuned_term_Ccre", header = 7)
```

```python
lis = [bsub, ecol, vnat[vnat["Chromosome"]==1], vnat[vnat["Chromosome"]==2], ccre]

for i in range(len(lis)):
    lis[i] = lis[i][["Position", "Strand", "Gene upstream", "Gene downstream", "Fraction of mRNA of downstream gene from readthrough"]].copy()

```

```python
lis[0]
```

```python
genes = ["NC_000964.3", "NC_000913.3", "CP009977.1", "CP009978.1", "CP001340.1"]
for i in range(len(lis)):
    lis[i]["gene"] = [genes[i] for j in range(len(lis[i]))]

    lis[i]["start"] = lis[i]["Position"] - 1
    lis[i]["end"] = lis[i]["Position"]
    lis[i].drop(["Position"], axis = 1, inplace = True)

    lis[i]["Strand"] = lis[i]["Strand"].replace({1: "+", 0: "-"})

    lis[i]["Gene upstream"] = lis[i]["Gene upstream"].fillna("Unknown")
    lis[i]["Gene downstream"] = lis[i]["Gene downstream"].fillna("Unknown")
    lis[i]["name"] = lis[i]["Gene upstream"].str.strip() + "_" + lis[i]["Gene downstream"].str.strip()
    lis[i].drop(["Gene upstream", "Gene downstream"], axis = 1, inplace = True)

    lis[i] = lis[i][lis[i]["Fraction of mRNA of downstream gene from readthrough"]>=0.2]

    lis[i]["score"] = lis[i]["Fraction of mRNA of downstream gene from readthrough"]
    lis[i].drop(["Fraction of mRNA of downstream gene from readthrough"], axis = 1, inplace = True)

    # lis[i]["Closest upstream gene"] = lis[i]["name"].str.strip()

    lis[i] = lis[i][["gene", "start", "end", "name", "score", "Strand"]]

```

```python
lis[0].to_csv('BacillusSubtilis(ReadThrough)_168_NC000964.3.bed', sep='\t', index=False, header=False)
```

```python
lis[1].to_csv('EscherichiaColi(ReadThrough)_K12-MG1655_NC000913.3.bed', sep='\t', index=False, header=False)
```

```python
lis[2].to_csv('VibrioNatriegens(ReadThrough)_ATCC14048_CP009977.1.bed', sep='\t', index=False, header=False)
lis[3].to_csv('VibrioNatriegens(ReadThrough)_ATCC14048_CP009978.1.bed', sep='\t', index=False, header=False)
```
