```python
import pandas as pd
# show all columns 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel("Table_S5_Bastet_et_al.xlsx", sheet_name="TableS5", header=0, skipfooter=399)
df
```

```python
df = df[["Strand", "ΔG by Quickfold", "Start_TT", "Stop_TT", "TERM_ID"]]
df
```

```python
df["gene"] = [f"NC_007795.1_{i}" for i in range(len(df))]

df.drop(columns=["Start_TT"], inplace=True)
df
```

```python

df["start"] = df["Stop_TT"] - 1
df["end"] = df["Stop_TT"]


df.drop(columns=["Stop_TT"], inplace=True)
df
```

```python
df = df[["gene", "start", "end", "TERM_ID", "ΔG by Quickfold", "Strand"]]
df
```

```python
df.to_csv("StaphylococcusAureus_NCTC8325_NC007795.1.bed", sep='\t', header=False, index=False)
```
