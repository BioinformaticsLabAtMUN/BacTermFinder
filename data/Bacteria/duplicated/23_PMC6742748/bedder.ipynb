```python
import pandas as pd
```

```python
df = pd.read_excel("Table_6.xlsx", sheet_name= "Sheet1", header = 0)
df
```

```python
df = df[(df["Category"] == "P") | (df["Category"] == "S")]
df.drop(["Category",], axis = 1, inplace = True)
df
```

```python
df["gene"] = ["CP009124.1" + "_{}".format(j) for j in range(len(df))]
df
```

```python
df['start'] = df['Position'] - 1
df['end'] = df['Position']

df.drop(["Position"], axis = 1, inplace = True)

df
```

```python

df = df[["gene", "start", "end", "Associated gene", "Intensity", "Strand"]]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('StreptomycesLividans_TK24_CP009124.1.bed', sep='\t', index=False, header=False)
```
