```python
import pandas as pd
```

```python
df = pd.read_excel("mgen-7-524-s002.xlsx", sheet_name= "Terminators", header = 0)
```

```python
df.head ()
```

```python

df.drop(["Terminator"], axis = 1, inplace = True)
    
```

```python

df["gene"] = ["ASM904v1" + "_{}".format(j) for j in range(len(df))]
df
```

```python
fwd = df[df["strand"] == "+"]
rev = df[df["strand"] == "-"]

df["POT"] = pd.concat([fwd["end"], rev["start"]])
df
```

```python


df['start'] = df['POT'] - 1
df['end'] = df['POT']

df.drop(["POT"], axis = 1, inplace = True)

df
```

```python

df = df[["gene", "start", "end", "annotation origin", "Free energy [kcal/mol]", "strand"]]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('BacillusSubtilis_168_ASM904v1.bed', sep='\t', index=False, header=False)
```
