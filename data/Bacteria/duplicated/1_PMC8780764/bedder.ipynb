```python
import pandas as pd
```

```python
df = pd.read_excel('12864_2022_8314_MOESM4_ESM.xlsx', sheet_name='Supplementary Dataset S3', skiprows=1 )
df
```

```python
# filter results to only include the rows with the correct TEP category
df = df[df['TEP category'].isin(['P', 'S'])]
df
```

```python
# chromosome, start, end, Associated gene , score, strand
df = df.assign(chromosome = ["BA000030.4" for _ in range(len(df))])
df
```

```python
df['TEP position']- 1
```

```python
# substract 1 from "TEP position" column and add it to a new column
df = df.assign(start=df['TEP position']- 1)
df
```

```python
# Rename a column
df['end'] = df['TEP position'].copy()
df
```

```python
df['name'] = df["Associated gene"].copy()
df
```

```python
df['score'] = [0 for _ in range(len(df))]
df
```

```python
df['strand'] = df["Strand"].copy()
df
```

```python
df.drop(["TEP position", "Strand", "Signal intensity", "Associated gene", "TEP category"], axis=1, inplace=True)
```

```python
df
```

```python
# convert df to bed file format  
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed  
df.to_csv('StreptomycesAvermitilis_MA-4680_BA000030.4.bed', sep='\t', header=False, index=False)
```
