```python
import pandas as pd
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv('pcbi.0010025.sd001 (2).txt', sep='\t')
```

```python
df.head()
```

```python
# find row indexes that contain "b/Bidirectional" string in the column "Comments" 
backup = df[df['Comments'].str.contains('idirectional').replace(float("NaN"),False).astype(bool)] 
backup_idx = backup.index
```

```python

df = pd.concat([df, backup], ignore_index=True)
df['strand'] = ["+" for _ in range (len(df))]
df
```

```python
df.loc[backup_idx, 'strand'] = "-"
```

```python

df.strand.value_counts()
```

```python
df.drop(["Genes", "Terminator", "Evidence", "References", "Comments"], axis=1, inplace=True)
df
```

```python
df["gene"] = [f"NC_000964" for _ in range(len(df))]
# df["strand"] = [" " for _ in range(len(df))]
df.head()
```

```python
df["Terminator position"].replace("ND",float("NaN"), inplace=True)

df.dropna(subset=["Terminator position"], inplace=True)

df
```

```python
df["start"] = df["Terminator position"].map(lambda x: x.split("..")[0])
df["end"] = df["Terminator position"].map(lambda x: x.split("..")[1])
df.drop(["Terminator position"], axis=1, inplace=True)
df
```

```python
df.reset_index(drop=True, inplace=True)
df["gene"] = ["NC_000964.3" for _ in range(len(df["gene"]))]
df
```

```python
# replace "ND" with NaN for all dataframe
df.replace("ND", float("NaN"), inplace=True)
df.isna().sum()
```

```python
# drop rows with NaN only in the "Terminator position" column
df.dropna(subset=["Terminator position"], inplace=True)
df.isna().sum()
```

```python
# ffill the DeltaG values with the mean of the row
df["Delta G [kcal/mole]"] = df["Delta G [kcal/mole]"].fillna(df.groupby("gene")["Delta G [kcal/mole]"].transform("mean"))

# basically filled them out with 0 :(
```

```python
df.replace(float("NaN"), 0, inplace=True)
df.isna().sum()
```

```python
fw = df['strand'] == "+"
bw = df['strand'] == "-"

fw
```

```python
df
```

```python

# concat two pd series a, b
idx = pd.concat([df[fw]["end"] ,df[bw]["start"]], ignore_index=True)
idx.sort_index(inplace=True)
idx
```

```python
df["tts"] = idx
df.drop(["start", "end"], axis=1, inplace=True)
```

```python


df["start"] = df["tts"].astype(int) - 1
df["end"] = df["tts"].astype(int)

df.drop(["tts"], axis=1, inplace=True)

df
```

```python
# change order of the columns   
df = df[['gene', 'start', "end", 'Operon name', 'Delta G [kcal/mole]', 'strand']].copy()
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('BacillusSubtilis_168_NC000964.3.bed', sep='\t', header=False, index=False)
```
