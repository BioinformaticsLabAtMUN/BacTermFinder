```python
import pandas as pd
```

```python
df = pd.read_csv("GSE67058_terminators.txt", delimiter="\t")
df
```

```python
df.drop(["Relative position", "%T (+ NusA)", "%T (– NusA)", "p-value", "Terminator identified", "U-tract (9 nt)", "Hairpin mismatch"], axis=1, inplace=True)
df
```

```python
print(df.isna().sum())
```

```python
df["gene"] = [f"NC_000964.3_{i}" for i in range(len(df))]
df
```

```python
df["start"] = df["POT"] - 1
df["end"] = df["POT"]
df
```

```python
df = df[["gene", "start", "end", "Δ%T", "ΔG (kCal/mol)", "Strand"]]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('BacillusSubtilis_168_NC000964.3.bed', sep='\t', index=False, header=False)
```
