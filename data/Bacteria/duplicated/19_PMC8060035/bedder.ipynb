```python
import pandas as pd
```

```python
df = pd.read_excel("elife-61880-supp2.xlsx", sheet_name= "NusA depletion", header = 0)

df
```

```python
df = df[["POT", "Strand", "ΔG-Hairpin (kcal/mol)", "Relative Position"]]

df
```

```python
df["gene"] = ["NC_000964.3_{}".format(i) for i in range(len(df))]
df
```

```python
df['start'] = df['POT'] - 1
df['end'] = df['POT']

df.drop(["POT"], axis = 1, inplace = True)

df
```

```python
df = df [["gene", "start", "end", "Relative Position", "ΔG-Hairpin (kcal/mol)", "Strand"]]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('BacillusSubtilis_168_NC000964.3.bed', sep='\t', index=False, header=False)

```
