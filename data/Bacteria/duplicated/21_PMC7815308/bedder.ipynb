```python
import pandas as pd
```

```python
lb4 = pd.read_excel("elife-62438-supp1_.xlsx", sheet_name= "LB 0.4", header = 1)
lb2 = pd.read_excel("elife-62438-supp1_.xlsx", sheet_name= "LB 2.0", header = 1)
m63 = pd.read_excel("elife-62438-supp1_.xlsx", sheet_name= "M63 0.4", header = 1)
```

```python
# append all of the dataframes into one big and merge them on the first column as index
df = pd.concat([lb4, lb2, m63], axis=0, ignore_index=True)
```

```python
df
```

```python
# select the row that have unique values of the first column
df = df.drop_duplicates(subset=['3´ end position'])
```

```python
df
```

```python
# sort the dataframe by the first column
df = df.sort_values(by=['3´ end position'] , ascending=True)
# reset the index
df = df.reset_index(drop=True)
```

```python
df
```

```python
df['classification'].value_counts()
```

```python
df.drop(["3′ end region",], axis = 1, inplace = True)
df
```

```python
# if the value in the classification column has orphan or antisense, then drop the row.
df = df[~df['classification'].isin(['orphan', 'antisense, ', 'primary, antisense, ', 'internal, antisense, '])]
df
```

```python
df["gene"] = ["NC_000913.3" + "_{}".format(j) for j in range(len(df))]
df
```

```python
df["name"] = df["details"] + " | " + df["classification"]
```

```python

df.drop(["details", "classification"], axis = 1, inplace = True)
df
```

```python
df['start'] = df['3´ end position'] - 1
df['end'] = df['3´ end position']

df.drop(["3´ end position"], axis = 1, inplace = True)

df
```

```python

df = df[["gene", "start", "end", "name", "Term-seq AVG read count", "strand"]]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('EscherichiaColi_K12-MG1655_NC000913.3.bed', sep='\t', index=False, header=False)
```
