```python
import pandas as pd
# show all columns 
pd.set_option('display.max_columns', None)
```

```python
bsub = pd.read_excel("NIHMS1597345-supplement-1597345_Supp_Data_Tab2.xlsx", sheet_name="Bsub_term", header=1, skipfooter=0)
ecoli = pd.read_excel("NIHMS1597345-supplement-1597345_Supp_Data_Tab2.xlsx", sheet_name="Ecoli_term", header=1, skipfooter=0)
print(len(ecoli))
print(len(bsub))
```

```python
bsub = bsub[["Position", "Strand", "Gene upstream", "Readthrough fraction (WT)"]]
ecoli = ecoli[["Position", "Strand", "Gene upstream", "Readthrough fraction (WT)"]]

bsub['start'] = bsub['Position'] - 1
ecoli['start'] = ecoli['Position'] - 1

bsub['end'] = bsub['Position']
ecoli['end'] = ecoli['Position']

bsub.drop(columns=['Position'], inplace=True)
ecoli.drop(columns=['Position'], inplace=True)

# To do
bsub['gene'] = [f"NC_000964.3_{i}" for i in range(len(bsub))]
ecoli['gene'] = [f"NC_000913.3_{i}" for i in range(len(ecoli))]
```

```python
bsub["Strand"] = bsub["Strand"].apply(lambda x: '+' if x==1 else '-')
```

```python
ecoli["Strand"] = ecoli["Strand"].apply(lambda x: '+' if x==1 else '-')
```
