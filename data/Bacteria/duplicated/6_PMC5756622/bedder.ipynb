```python
import pandas as pd
```

```python
df = pd.read_excel("NIHMS75525-supplement-Table_S1.xlsx")

df
```

```python
# chromosome, start, end, Associated gene , score, strand
df['gene'] = ["AL009126.3" for _ in range(len(df))]
df
```

```python
# drop 4th column in dataframe by index
df.drop(columns=["Gene locus", "Sequence (40nt) upstream of Term-seq position", "Upstream sequence (40nt) fold (RNAfold)", "# of Uridines found in the 8nt immediately upstream to the Term-seq site"], axis=1, inplace=True)
df
```

```python
df["start"] = df["Term-seq position"] - 1
df["end"] = df["Term-seq position"]
df["strand"] = [" " for _ in range(len(df))]
df
```

```python
org = pd.read_excel("NIHMS75525-supplement-Table_S1.xlsx")
df["name"] = org["Gene locus"].copy()
df
```

```python
df.drop(columns=["Term-seq position"], axis=1, inplace=True)
```

```python
# change order of the columns   
df = df[['gene', 'start', 'end', 'name', 'Folding energy (kcal/mol)', 'strand']]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('BacillusSubtilis_168_AL009126.3', sep='\t', header=False, index=False)
```

# Adding strand from gene annotation

```python
# read sequence.txt file and save the lines that have "complement" in them in a list
with open("sequence.txt", "r") as f:
    lines = f.readlines()
    lines = [line.strip() for line in lines if "complement" in line]
```

```python

# extract the locus tag of the complement regions
locus = ["BSU" + line.split("locus_tag=BSU_")[1].split("]")[0] for line in lines]
print(locus[1:10])
```

```python
import pandas as pd
term = pd.read_csv("BacillusSubtilis_168_AL009126.3.bed", sep='\t', header=None)
term.head()
```

```python
term[5] = term.apply(lambda x: "+" if x[3] not in locus else "-", axis=1)
```
