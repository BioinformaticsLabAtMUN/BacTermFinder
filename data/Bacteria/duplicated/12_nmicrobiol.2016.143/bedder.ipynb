```python
import pandas as pd
```

```python
s4 = pd.read_excel("41564_2016_BFnmicrobiol2016143_MOESM264_ESM.xlsx", sheet_name="Table S4")
s5 = pd.read_excel("41564_2016_BFnmicrobiol2016143_MOESM264_ESM.xlsx", sheet_name="Table S5")
```

```python
# append s5 to s4 using panadas concat
df = pd.concat([s4, s5], axis=0)
df
```

```python
# select only the columns we need
df = df[["Locus tag", "Gene strand", "Terminator position", "Fold stability", ]]
```

```python
df["start"] = df["Terminator position"].astype(int) - 1
df["end"] = df["Terminator position"].astype(int) 

df.drop(columns=["Terminator position"], inplace=True)

df
```

```python
df["gene"] = [f"AL009126.3_{i}" for i in range(len(df))]

# df.drop(columns=[""], inplace=True)
df = df [["gene", "start", "end", "Locus tag", "Fold stability", "Gene strand"]]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    

df.to_csv('BacillusSubtilis_168_AL009126.3.bed', sep='\t', index=False, header=False)
```
