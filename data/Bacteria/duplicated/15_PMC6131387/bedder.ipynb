```python
import pandas as pd
```

```python
m9 = pd.read_excel("41467_2018_5997_MOESM4_ESM.xlsx", sheet_name= "M9_TTS", header = 1)
rich = pd.read_excel("41467_2018_5997_MOESM4_ESM.xlsx", sheet_name= "Rich_TTS", header = 1)
m9
```

```python
rich
```

```python
m9.drop(m9.columns[-7:-1], axis=1, inplace=True)
rich.drop(rich.columns[-7:-1], axis=1, inplace=True)

m9
```

```python
m9.drop(m9.columns[4::2], axis=1, inplace=True)
rich.drop(rich.columns[4::2], axis=1, inplace=True)
m9
```

```python
m9["name"] = "m9"
rich["name"] = "rich"
```

```python
m9["genome"] = ["NC_000913.3_{}".format(i) for i in m9.index]

rich["genome"] = ["NC_000913.3_{}".format(i) for i in rich.index]

m9
```

```python
m9.drop(["# genome"], axis=1, inplace=True)
```

```python
fw_m9 = m9.loc[m9["strand"] == "forward"]
rv_m9 = m9.loc[m9["strand"] == "reverse"]
m9["tts"] = pd.concat([fw_m9["end"] ,  rv_m9["start"]]).sort_index()
m9.drop(["start", "end"], axis=1, inplace=True)

fw_rich = rich.loc[rich["strand"] == "forward"]
rv_rich = rich.loc[rich["strand"] == "reverse"]
rich["tts"] = pd.concat([fw_rich["end"] ,  rv_rich["start"]]).sort_index()
rich.drop(["start", "end"], axis=1, inplace=True)

print(m9.head())
print(rich.head())
```

```python

m9["strand"] = m9["strand"].map({"forward": "+", "reverse": "-"})

m9
```

```python
rich["strand"] = rich["strand"].map({"forward": "+", "reverse": "-"})
rich
```

```python
m9["start"] = m9["tts"] - 1
m9["end"] = m9["tts"]

rich["start"] = rich["tts"] - 1
rich["end"] = rich["tts"]

m9.drop(["tts"], axis=1, inplace=True)
rich.drop(["tts"], axis=1, inplace=True)
```

```python
m9 = m9 [["genome", "start", "end", "name", "number_of_reads_TTS", "strand"]]
rich = rich [["genome", "start", "end", "name", "number_of_reads_TTS", "strand"]]

print(m9.head())
print(rich.head())
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    

m9.to_csv('EscherichiaColi_K12-MG1655(m9)_NC000913.3.bed', sep='\t', index=False, header=False)
rich.to_csv('EscherichiaColi_K12-MG1655(rich)_NC000913.3.bed', sep='\t', index=False, header=False)
```
