```python
import pandas as pd
```

```python
df = pd.read_csv("41564_2019_500_MOESM4_ESM.csv")
df
```

```python
# chromosome, start, end, Associated gene , score, strand
df['TTS_serial_number'] = [f"NC_000913.3_{i}" for i in range(len(df))]
df
```

```python
# drop 4th column in dataframe by index
df.drop(df.columns[4:6], axis=1, inplace=True)
```

```python
df
```

```python
df["start"] = df['TTS_position']- 1
df
```

```python
# map "intrinsic_TTS" and rho dep one to binary
mapper = {float("nan"): 0, "rho-dependent": 2,}
df["TTS_type_rho_dependent"] = df["TTS_type_rho_dependent"].map(mapper)

mapper = {float("nan"): 0, "intrinsic_TTS": 1,}
df["intrinsic_TTS"] = df["intrinsic_TTS"].map(mapper)

df
```

```python
# combine 2 columns into 1
df["name"] = df["intrinsic_TTS"] + df["TTS_type_rho_dependent"]
df
```

```python
df.drop(["intrinsic_TTS", "TTS_type_rho_dependent"], axis=1, inplace=True)
```

```python
df
```

```python
mapper = {0:"undefined", 1:"intrinsic", 2:"rho-dependent"}
df["name"] = df["name"].map(mapper)
df
```

```python
df["score"] = df[" ∆G (kcal/mol)"].copy()
df
```

```python
# change order of the columns   
df = df[['TTS_serial_number', 'start', 'TTS_position', 'name', 'score', 'TTS_strand']]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('EscherichiaColi_K12-MG1655_NC000913.3.bed', sep='\t', header=False, index=False)
```
