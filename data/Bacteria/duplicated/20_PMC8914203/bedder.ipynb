```python
import pandas as pd
```

```python
df = pd.read_excel("clean.xlsx", sheet_name= "Sheet1", header = 0)
df
```

```python
df.drop(["TEP ID",], axis = 1, inplace = True)
df
    
```

```python
df["gene"] = ["NC_010572.1" + "_{}".format(j) for j in range(len(df))]
df
```

```python
df["name"] = df["Gene"] + " | " + df["Subcategory"]
df.drop(["Gene", "Subcategory"], axis = 1, inplace = True)
df
```

```python
df = df[(df["Category"] == "P") | (df["Category"] == "S")]
```

```python
df['start'] = df['TEP position'] - 1
df['end'] = df['TEP position']

df.drop(["TEP position"], axis = 1, inplace = True)

df
```

```python

df = df[["gene", "start", "end", "name", "Abundance", "Strand"]]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('StreptomycesGriseus_NBRC13350_NC010572.1.bed', sep='\t', index=False, header=False)
```
