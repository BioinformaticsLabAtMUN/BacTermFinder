```python
import pandas as pd
```

```python
df = pd.read_excel("Supplementary Table S4.rev.xlsx", sheet_name= "Sheet1", header = 1)
df
```

```python

df.drop(["ID", "Location", "Upstream DNA sequence", "Dar (2018)", "Adams (2021)"], axis = 1, inplace = True)
df
    
```

```python
df["gene"] = ["NC_000913.3" + "_{}".format(j) for j in range(len(df))]
df
```

```python
df["name"] = df["Class"] + "_" + df["Genetic context"]
df.drop(["Class", "Genetic context"], axis = 1, inplace = True)
df
```

```python


df['start'] = df['Position'] - 1
df['end'] = df['Position']

df.drop(["Position"], axis = 1, inplace = True)

df
```

```python
df["score"] = 0
df = df[["gene", "start", "end", "name", "score", "Strand"]]
df
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    
df.to_csv('EscherichiaColi_K12-MG1655_NC000913.3.bed', sep='\t', index=False, header=False)
```
