```python
import pandas as pd
```

```python
df = pd.read_excel("msystems.00943-21-st005.xlsx", header=1)
df
```

```python
df = df[(df["Category"] == " P") | (df["Category"] == " S")].copy()
df
```

```python
df = df[["Position", "Strand", "Category", "Associated gene ID"]]
df
```

```python
df["gene"] = [f"NC_000911.1_{i}" for i in range(len(df))]
df
```

```python
df["start"] = df["Position"] - 1
df["end"] = df["Position"]
df["score"] = 0
df
```

```python
df = df[["gene", "start", "end", "Associated gene ID", "score", "Strand"]]
df
```

```python
df.isna().sum()
```

```python
# convert df to bed file format
# {FullNameOfTheBacteria}_{strainCode}_{GenomeAccessionNumber}.bed    

df.to_csv('Synechocystis_PCC6803_NC000911.1.bed', sep='\t', index=False, header=False)
```
