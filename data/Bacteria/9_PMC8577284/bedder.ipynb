```python
import pandas as pd
```

```python
df = pd.read_csv("TTSs.tab", sep="\t", header=None)
df.head()
```

```python
# chromosome, start, end, Associated gene , score, strand
df['gene'] = [f"NC_004603.1_{i}" for i in range(len(df))]
df
```

```python
df
```

```python
df.drop([1], axis=1, inplace=True)
df
```

```python
# change order of the columns   
df = df[['gene', 'start', 'end', 0, 2,]]
df
```

# Adding strand from gene annotation

```python
# read sequence.txt file and save the lines that have "complement" in them in a list
with open("sequence (2).txt", "r") as f:
    lines = f.readlines()
    lines = [line.strip() for line in lines if "complement" in line]
lines[:5]
```

```python

# extract the locus tag of the complement regions
locus = [line.split("locus_tag=")[1].split("]")[0] for line in lines]
print(locus[:5])
```

```python
df
```

```python
df[5] = df.apply(lambda x: "+" if x[0] not in locus else "-", axis=1)
df
```

```python
df.head(40)
```

```python
df.to_csv("VibrioParahaemolyticus_RIMD2210633_NC_004603.1.bed", sep='\t', header=False, index=False)
```
